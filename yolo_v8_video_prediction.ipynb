```python
# importing important dependencies
from ultralytics import YOLO             # YOLOv8 model
import cv2                               # opencv
import numpy as np                       # for mathematical operations
import random                            # to generate random numbers
```

```python
model1 = YOLO('yolov8n.pt')              # importing yolov8n pretrained model
```

```python
results = model1('video.mp4',stream = True)
```

```python
for result in results:
    boxes = result.boxes          # Boxes object for bbox outputs
    masks = result.masks          # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs          # Probs object for classification outputs
```

```python
my_file = open("coco.txt", "r")   # text file containing class name
data = my_file.read()
class_list = data.split("\n")     # getting the class names in form of a list
my_file.close()
```

```python
# creating random colors for each class to use in bounding box
detection_colors = []
for i in range(len(class_list)):
    r = random.randint(0, 255)          # value of red pixel
    g = random.randint(0, 255)          # value of green pixel
    b = random.randint(0, 255)          # value of blue pixel
    detection_colors.append((b, g, r))
```

```python
cap = cv2.VideoCapture('video.mp4')     # loaading the video   
while True:
    ret,frame = cap.read()              # reading the video frame by frame
    if ret == False:                    # if the video is read correctly then ret returns true
        print('unable to read video')
        break
    detect_params = model1.predict(source =[frame],conf=0.45, save=False) # using the model to predict the parameters of each frame
    for i in range(len(detect_params[0])):
        boxes = detect_params[0].boxes      # bounding box information
        box = boxes[i]                      # returns one box
        clsID = box.cls.numpy()[0]          # classIDs in the box 
        conf = box.conf.numpy()[0]          # confidence level of each class
        bb = box.xyxy.numpy()[0]            # coordinates of of bounding box (left,top,right,bottom)
        
        cv2.rectangle(frame,(int(bb[0]), int(bb[1])-30),(int(bb[2]), int(bb[1])),(0,0,0),-1) # rectangle to display text
        cv2.rectangle(frame,
                      (int(bb[0]), int(bb[1])),(int(bb[2]), int(bb[3])),
                      detection_colors[int(clsID)],3,)                       # bounding box rectangle
        
        
        font = cv2.FONT_HERSHEY_COMPLEX_SMALL                                # font formatting
        cv2.putText(frame,class_list[int(clsID)] +                           # text formating
                    " " + str(round(conf, 2)) + 
                    "%",(int(bb[0]), int(bb[1]) - 5),
                    font,1,(255, 255, 255),2,)
    cv2.imshow("ObjectDetection", frame)                                     # displaying each frame
    if cv2.waitKey(1) == 27:                                                 # using escape key to close the window
        break
cap.release()
cv2.destroyAllWindows()
```
